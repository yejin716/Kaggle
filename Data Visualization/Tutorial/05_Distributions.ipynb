```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
path = '../../../data/iris.csv'
df = pd.read_csv(path, index_col='Id')

df.head()
```

```python
#Histograms #도수분포를 정보그림으로 나타낸것
sns.histplot(df['Petal Length (cm)'])
```

```python
#Kdeplot >> 커널밀도추정 플롯 

sns.kdeplot(data=df['Petal Length (cm)'], shade=True)
#shade=True >> 곡선아래의 영역을 색으로 채울것인가
```

```python
# 2D KDE plot
sns.jointplot(x = df['Petal Length (cm)'], y=df['Sepal Width (cm)'], kind='kde')
```

```python
#색상 추가 
sns.histplot(data=df, x='Petal Length (cm)', hue='Species')
```

```python
#kdeplot에 각 종마다 색상 지정 
sns.kdeplot(data=df, x ='Petal Length (cm)', hue='Species', shade=True)
```
