```python
import pandas as pd 

reviews = pd.read_csv('../../data/winemag-data-130k-v2.csv', index_col=0)
```

```python
#grouping 

#points열을 그룹화하고 각 값의 개수 반환
reviews.groupby('points')['points'].count()
```

```python
reviews.groupby('points')['price'].min()
#points별로 price의 가장 작은 가격 반환 
```

```python
reviews['title'].iloc[0]
```

```python
reviews.groupby('winery').apply(lambda df : df['title'].iloc[0])
#winery별로 나오는 title의 첫번째 값을 적용 
```

```python
reviews.groupby(['country','province']).apply(lambda df: df.loc[df['points'].idxmin()])
#country와 province별로 나눴을때 points의 값이 큰 인덱스에 있는 데이터를 불러오는것 
```

```python
reviews.groupby(['country'])['price'].agg([len,min,max])
#각 나라별 가격의 길이, 최소값, 최대값 나타내기 
#aggregate >집계연산
```

```python
#그룹화를 하면 인덱스가 다중으로 들어갈 수 있다. 
countries_reviewed = reviews.groupby(['country', 'province'])['description'].agg([len])
countries_reviewed
```

```python
mi = countries_reviewed.index
type(mi)

#다중인덱스 
```

```python
countries_reviewed.reset_index()
#다중인덱스를 열로 보내기 
```

```python
#sorting 정렬 

countries_reviewed = countries_reviewed.reset_index()
countries_reviewed.sort_values(by='len')
#len기준으로 오름차순 정렬
```

```python
countries_reviewed.sort_values(by='len', ascending=False)
#ascending=False >> 내림차순
```

```python
countries_reviewed.sort_index()
#index기준으로 오름차순
```

```python
countries_reviewed.sort_values(by=['country','len'])
#2개이상의 열을 정렬할 수 있음 
```
